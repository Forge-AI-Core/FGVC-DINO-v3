```python
from pathlib import Path
from typing import Any
import json


annotation_path = Path("/home/jongha/projects/FGVC-DINO-v3/data/Iron-Scraps/my_classification/annotations.json")


def load_coco_annotations(annotation_path: Path) -> dict[str, Any]:
    with open(annotation_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)

    return json_data

json_data = load_coco_annotations(annotation_path=annotation_path)
```

```python
json_items = json_data.items()
for key, _ in json_items:
    print(key)
```
