```python
import torch
import yaml

with open(file="hyperparams.yaml", mode="r", encoding="utf-8") as f:
    hyperparams = yaml.safe_load(f)


model_name = hyperparams["model"]["vits16"]["NAME"]
model_path = hyperparams["model"]["vits16"]["PATH"]


model = torch.hub.load(
        repo_or_dir="dinov3",
        model=model_name,
        source="local",
        weights=str(model_path),
    )
```
